**DATA PREPROCESSING**

```python
df['question1']=df['question1'].str.lower()
df['question2']=df['question2'].str.lower()
```

```python
df['question1']=df['question1'].str.strip()
df['question2']=df['question2'].str.strip()
```

```python
df['question1']=df['question1'].replace('%','percent')
df['question1']=df['question1'].replace('$','dollar')
df['question1']=df['question1'].replace('@','at')
df['question1']=df['question1'].replace('₹','rupees')
df['question1']=df['question1'].replace('€','euro')
```

```python
df['question2']=df['question2'].replace('%','percent')
df['question2']=df['question2'].replace('$','dollar')
df['question2']=df['question2'].replace('@','at')
df['question2']=df['question2'].replace('₹','rupees')
df['question2']=df['question2'].replace('€','euro')
```

```python
import re
pattern=re.compile(r'\W')
df['question1'] = df['question1'].apply(lambda x: re.sub(pattern, '', str(x)).strip() if isinstance(x, str) else x)
```

```python
df['question1']=df['question1'].replace('[math]','')
df['question2']=df['question2'].replace('[math]','')
```

**TRAIN-TEST SPLIT**

```python
X=df.iloc[: ,3:5]
y=df.iloc[: ,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=42)
```

**Tokenization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=3000)
```

```python
X_train['question1'] = X_train['question1'].fillna('')
```

```python
X_train_cv=cv.fit_transform(X_train['question1']).toarray()
X_test_cv=cv.transform(X_test['question1']).toarray()
```

```python
X_train_cv.shape
```

```python
y_train.shape[0]
```

**MODEL TRAINING**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10)
rfc.fit(X_train_cv,y_train)
```

```python
y_pred=rfc.predict(X_test_cv)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train_cv,y_train)
y_pred1 = xgb.predict(X_test_cv)
accuracy_score(y_test,y_pred1)
```
